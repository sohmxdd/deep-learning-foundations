## What is a Perceptron?

A perceptron is a basic building block of neural networks.
It computes a weighted sum of inputs, adds a bias, and applies a step function to produce a binary output.

```python
import numpy as np
#These are our input features
X= np.array([1.0,2.0])
#These are our weights
W= np.array([0.5,0.5])
#Bias
b=0.1
```

```python
#Calculating z
z=np.dot(X,W)+b
```

```python
#Step activation function
def step_function(z):
    if z>=0:
        return 1
    else:
        return 0
#Getting the output
output=step_function(z)
```

```python
output
```

## Limitations of a Perceptron

- Can only classify linearly separable data
- Cannot solve non-linear problems like XOR
- Uses a hard threshold (step function), which is not differentiable

Because of these limitations, perceptrons are mainly used for understanding
the foundations of neural networks rather than real-world applications.

### Observation

Changing the weights or bias shifts the decision boundary and can change
the final prediction of the perceptron.
