Comparing common regression loss functions on the same predictions.

```python
import numpy as np

# Ground truth values
y_true = np.array([3, 5, 2.5, 7])

# Model predictions
y_pred = np.array([2.5, 5, 4, 8])

# Mean Squared Error
mse = np.mean((y_true - y_pred) ** 2)

# Mean Absolute Error
mae = np.mean(np.abs(y_true - y_pred))

mse, mae
```
